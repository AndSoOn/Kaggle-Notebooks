```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#from Ipython.display import display
%matplotlib inline
sns.set(style='white', context='notebook', palette='deep')
mycols = ["#66c2ff", "#5cd6d6", "#00cc99", "#85e085", "#ffd966", "#ffb366", "#ffb3b3", "#dab3ff", "#c2c2d6"]
sns.set_palette(palette = mycols, n_colors = 4)
%config InlineBackend.figure_format = 'retina'
```

```python
path = "/Users/ZHHR/Desktop/python_codes/git_all/Kaggle/competitions/\
House Prices: Advanced Regression Techniques/"
```

```python
df_train = pd.read_csv(f'{path}train.csv', index_col="Id")
df_test = pd.read_csv(f'{path}test.csv', index_col="Id")
print(df_train.shape)
print(df_test.shape)
df_train.head()
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.regplot(x=df_train.GrLivArea, y=df_train.SalePrice, fit_reg=True).set_title("Before")
plt.subplot(1, 2, 2)
expl = df_train[df_train.GrLivArea<=4000]
sns.regplot(x=expl.GrLivArea, y=expl.SalePrice, fit_reg=True).set_title("After")
```

```python
df_train = df_train.drop(df_train[df_train.GrLivArea>4000].index)
ntrain, ntest = df_train.shape[0], df_test.shape[0]
ytrain = df_train.SalePrice.values
df_all = pd.concat([df_train, df_test])
df_all.reset_index(drop=True)
df_all.drop(['SalePrice'], axis=1, inplace=True)
df_all.shape
```

```python
all_na = df_all.isnull().sum()
all_na = all_na[all_na != 0].sort_values(ascending=False)
plt.subplots(figsize=(15, 6))
all_na.plot(kind='bar')
```

```python
plt.subplots(figsize =(15, 5))

plt.subplot(1, 2, 1)
g = sns.countplot(x = "Utilities", data = df_train).set_title("Utilities - Training")

plt.subplot(1, 2, 2)
g = sns.countplot(x = "Utilities", data = df_test).set_title("Utilities - Test")
```

```python
cols1 = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageQual', 
         'GarageCond', 'GarageFinish', 'GarageType', 'BsmtExposure', 'BsmtCond', 
         'BsmtQual', 'BsmtFinType2', 'BsmtFinType1', 'MasVnrType']
for col in cols1:
    df_all[col] = df_all[col].fillna('None')

cols2 = ['MasVnrArea', 'BsmtFullBath', 'BsmtHalfBath', 'BsmtUnfSF', 'TotalBsmtSF', 
         'GarageYrBlt', 'GarageArea', 'GarageCars', 'BsmtFinSF2', 'BsmtFinSF1']
for col in cols2:
    df_all[col] = df_all[col].fillna(0)


df_all['LotFrontage'] = df_all.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

cols3 = ['MSZoning', 'Functional', 'SaleType', 
         'KitchenQual', 'Exterior2nd', 'Exterior1st', 'Electrical']
for col in cols3:
    df_all[col] = df_all[col].fillna(df_all[col].mode()[0])

all_na = df_all.isnull().sum()
all_na = all_na[all_na != 0].sort_values(ascending=False)

df_all = df_all.drop(['Utilities'], axis=1)
```

```python
corr = df_train.corr()

plt.subplots(figsize=(25, 25))
cmap = sns.diverging_palette(150, 250, as_cmap=True)
sns.heatmap(corr, cmap="RdYlBu", vmax=1, vmin=-0.6, center=0.2, 
            square=True, linewidths=0, cbar_kws={"shrink": .5}, annot = True);
```

```python
# Quadratic
df_all["OverallQual-2"] = df_all["OverallQual"] ** 2
df_all["GrLivArea-2"] = df_all["GrLivArea"] ** 2
df_all["GarageCars-2"] = df_all["GarageCars"] ** 2
df_all["GarageArea-2"] = df_all["GarageArea"] ** 2
df_all["TotalBsmtSF-2"] = df_all["TotalBsmtSF"] ** 2
df_all["1stFlrSF-2"] = df_all["1stFlrSF"] ** 2
df_all["FullBath-2"] = df_all["FullBath"] ** 2
df_all["TotRmsAbvGrd-2"] = df_all["TotRmsAbvGrd"] ** 2
df_all["Fireplaces-2"] = df_all["Fireplaces"] ** 2
df_all["MasVnrArea-2"] = df_all["MasVnrArea"] ** 2
df_all["BsmtFinSF1-2"] = df_all["BsmtFinSF1"] ** 2
df_all["LotFrontage-2"] = df_all["LotFrontage"] ** 2
df_all["WoodDeckSF-2"] = df_all["WoodDeckSF"] ** 2
df_all["OpenPorchSF-2"] = df_all["OpenPorchSF"] ** 2
df_all["2ndFlrSF-2"] = df_all["2ndFlrSF"] ** 2
print("Quadratics done!...")

# Cubic
df_all["OverallQual-3"] = df_all["OverallQual"] ** 3
df_all["GrLivArea-3"] = df_all["GrLivArea"] ** 3
df_all["GarageCars-3"] = df_all["GarageCars"] ** 3
df_all["GarageArea-3"] = df_all["GarageArea"] ** 3
df_all["TotalBsmtSF-3"] = df_all["TotalBsmtSF"] ** 3
df_all["1stFlrSF-3"] = df_all["1stFlrSF"] ** 3
df_all["FullBath-3"] = df_all["FullBath"] ** 3
df_all["TotRmsAbvGrd-3"] = df_all["TotRmsAbvGrd"] ** 3
df_all["Fireplaces-3"] = df_all["Fireplaces"] ** 3
df_all["MasVnrArea-3"] = df_all["MasVnrArea"] ** 3
df_all["BsmtFinSF1-3"] = df_all["BsmtFinSF1"] ** 3
df_all["LotFrontage-3"] = df_all["LotFrontage"] ** 3
df_all["WoodDeckSF-3"] = df_all["WoodDeckSF"] ** 3
df_all["OpenPorchSF-3"] = df_all["OpenPorchSF"] ** 3
df_all["2ndFlrSF-3"] = df_all["2ndFlrSF"] ** 3
print("Cubics done!...")

# Square Root
df_all["OverallQual-Sq"] = np.sqrt(df_all["OverallQual"])
df_all["GrLivArea-Sq"] = np.sqrt(df_all["GrLivArea"])
df_all["GarageCars-Sq"] = np.sqrt(df_all["GarageCars"])
df_all["GarageArea-Sq"] = np.sqrt(df_all["GarageArea"])
df_all["TotalBsmtSF-Sq"] = np.sqrt(df_all["TotalBsmtSF"])
df_all["1stFlrSF-Sq"] = np.sqrt(df_all["1stFlrSF"])
df_all["FullBath-Sq"] = np.sqrt(df_all["FullBath"])
df_all["TotRmsAbvGrd-Sq"] = np.sqrt(df_all["TotRmsAbvGrd"])
df_all["Fireplaces-Sq"] = np.sqrt(df_all["Fireplaces"])
df_all["MasVnrArea-Sq"] = np.sqrt(df_all["MasVnrArea"])
df_all["BsmtFinSF1-Sq"] = np.sqrt(df_all["BsmtFinSF1"])
df_all["LotFrontage-Sq"] = np.sqrt(df_all["LotFrontage"])
df_all["WoodDeckSF-Sq"] = np.sqrt(df_all["WoodDeckSF"])
df_all["OpenPorchSF-Sq"] = np.sqrt(df_all["OpenPorchSF"])
df_all["2ndFlrSF-Sq"] = np.sqrt(df_all["2ndFlrSF"])
print("Roots done!...")
```
